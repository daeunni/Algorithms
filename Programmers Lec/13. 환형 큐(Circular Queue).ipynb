# 13. 환형 큐(Circular Queue)

## 1) 큐(Queue)의 활용
- 자료를 생성하는 작업과 그 자료를 이용하는 작업이 비동기적으로(asynchronously)일어나는 경우 (하나 만들어서 하나 사용하고 이런 느낌)        


- 자료를 이용하는 작업이 여러곳에서 일어나는 경우     


- 자료를 처리하여 새로운 자료를 생성하고, 나중에 그 자료를 또 처리해야하는 작업의 경우   


-----

## 2) 환형 큐(Circular Queue)
- 정해진 개수의 저장공간을 빙 돌려가며 이용
> dequeue할 때 연산 복잡도를 줄이기 위해     
(이전 강의에서 배웠던 단순 queue와의 차이점!)

- 큐가 가득차면 더이상 원소를 넣을 수 없기 때문에 큐 길이를 기억하고 있어야 함
> isFull() : 큐의 원소가 꽉 차있는지를 판단하는 연산

- **rear(집어넣는 부분)**와 **front(꺼내는부분)** 포인터를 가지게 함으로써 돌려가면서 이용
> front와 rear를 적절히 계산해 배열을 환형으로 재활용



-----

## 3) 배열로 구현한 환형 큐

-----
#### < 주의할 점 > 
rear, front 포인터를 지정할 때는 꼭 % maxCount를 해줘야한다!!!!!!    
왜냐면 이게 환형 큐이기 때문에 index가 6 이하로 되게 지정해줘야하기 때문이다!

-------

```python
class CircularQueue:
    def __init__(self, n):
        self.maxCount = n    # 인자로 주어진 최대 큐 길이
        self.data = [None]*n  # 실제로 데이터를 저장할 공간 (우선 None 값을 채워서 n개 넣음)
        self.count = 0   # 현재 들어있는 원소의 개수
        self.front = -1  
        self.rear = -1   # 우선 -1로 해놓음(그 이유는??!?)
        
        
# 환형큐 구현        
def CircularQueue:
    
    def size(self):
        return self.count
    
    def isFull(self):
        return self.count == self.maxCount  # 큐가 꽉 차있는지 판별
    
    def enqueue(self, x):
        # 더이상 원소를 추가할 수 없으면
        if self.isFull():
            raise IndexError('Queue full')   # exception 처리
            
        self.rear = (self.rear+1) % self.maxCount  # rear 포인터 계산
        self.data[self.rear] = x
        self.count+=1
    
    def dequeue(self):
        # 큐가 비어있을 때
        if self.size()==0:
            raise IndexError('Queue empty')
        
        self.front = (self.front+1) % self.maxCount  # front 포인터 계산
        x = self.data[self.front]  # 데이터를 꺼내 x에 담는다
        self.count-=1
        return x  # 꺼낸 데이터를 return 
    
    def peek(self):
        if self.size()==0:
            raise IndexError('Queue empty')
            
        return self.data[(self.front+1) % self.maxCount]   # deque하면 튀어나올 첫 원소(처음 들어간게 나옴)
```
