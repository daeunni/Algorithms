### 2) 1부터 n까지 자연수의 합 구하기
- 재귀 알고리즘은 **종결 조건**이 매우 중요하다!

```python
def self_sum(n):
    return n + self_sum(n-1)   # 형식임
```

```python
def self_sum(n): 
    if n<= 1:      # 종결 조건이 매우 중요하다!
        return n 
    else:
        print(n)
        return n + self_sum(n-1)   # 형식임
    
self_sum(4)
```

- 단, 효율 측면에서 보면 조심해야한다 (함수를 호출해야하기 때문)

```python
# 펙토리얼 구하기

def fac(n):
    if n<=1:
        return 1
    else:
        return n* fac(n-1)
```

```python
# 피보나치 수열 (재귀 vs not 재귀)

def solution(x):   # x를 인덱스라고 생각
    if x<=1:
        return x
    else:
        return solution(x-1) + solution(x-2)
    
```
