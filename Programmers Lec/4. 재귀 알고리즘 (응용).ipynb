# 4. 재귀 알고리즘 응용
- 재귀 알고리즘은 함수를 호출해야하기 때문에 효율성 측면에서 한계가 있음   

### 1) 조합의 수 계산
n개의 서로다른 원소에서 m개를 택하는 경우


```python
# math 모듈 이용하기
from math import factorial as f

def combi(n, m):
    return f(n)/(f(m) * f(n-m))
```

< 재귀적인 방법으로 해결하기 >    

특정 원소를 포함하는 경우, 포함하지 않는 경우


```python
def combi(n, m):
    return combi(n-1, m) + combi(n-1, m-1)  

### 이건 틀림! trivial case를 고려하지 않았기 때문!

### 특정 조건에서는 trivial 하게 답을 리턴하는 코드 필요!
# >> if나 else로 지정하기
```

```python
def combi(n, m):
    if n==m :
        return 1
    elif m==0:
        return 1   # 이렇게 trivial 조건 
    else:
        return combi(n-1, m) + combi(n-1, m-1)    # 재귀함수
```

but 이는 효율성 측면에서 별로다! 함수를 여러번 호출하고있기 때문 



> fibo(1)을 두번  = **비효율성**(시간이 오래 걸린다)

#### 그럼 이렇게 효율이 떨어지는데 왜 재귀 알고리즘을 쓸까?   
> 사람이 생각하는 방식을 직접 코드로 옮길 수 있기 때문!   
ex) 하노이탑



```python
### 이진탐색의 재귀적 구현
# x의 위치를 선형 탐색이 아닌 이진 탐색으로 찾기! (정렬된 상황)

def binsearch(L, x, lower, upper):
    if x not in L:  
        return -1  # trivial 조건
    
    mid = (lower + upper)//2
    if x == L[mid]:
        return mid
    elif x < L[mid]:   # mid보다 작은 곳에서
        return binsearch(L, x, lower, mid-1)
    else:
        return binsearch(L, x, mid+1, upper)
        
    
```

### if x not in L:  이는 리스트 안에 x가 들어있는지 차례차례 탐색하는 과정으로, 선형 탐색 과정이다.
이는 O(n)의 복잡도가 소요되고, 코드가 실행은 되지만 비효율 적이다. 
따라서 주어진 부분을 다음과 같이 바꿔야한다

```python
### 이진탐색의 재귀적 구현
# x의 위치를 선형 탐색이 아닌 이진 탐색으로 찾기! (정렬된 상황)

def binsearch(L, x, lower, upper):
    if lower>upper :  
        return -1  # trivial 조건
    
    mid = (lower + upper)//2
    if x == L[mid]:
        return mid
    elif x < L[mid]:   # mid보다 작은 곳에서
        return binsearch(L, x, lower, mid-1)
    else:
        return binsearch(L, x, mid+1, upper)
```
