### 2) 리스트 순회
head에서 tail까지 끝까지 왔는가? 마지막 tail의 next는 None일 것임

```python
# 연결리스트의 노드 객체
class Node:
    def __init__(self, item):
        self.data = item  # 노드 안 데이터에 접근
        self.next = None  # 다음 노드에 접근하는 방법

# 연결리스트 객체
class LinkedList:
    def __init__(self):
        self.nodeCount = 0
        self.head = None  # 맨 처음 노드
        self.tail = None  # 맨 끝 노드

    # 특정 원소 참조 : pos 위치 curr로 반환
    def getAt(self, pos):
        if pos < 1 or pos > self.nodeCount:
            return None
        
        i = 1
        curr = self.head
        while i < pos:  # 선형 탐색 형식
            curr = curr.next
            i += 1
        return curr

# head에서 tail까지 끝까지 왔는가? tail의 next는 None으로 반환하면 종료
# 연결 노드 데이터를 리스트로 리턴
# 노드를 반환하는게 아니라 안에 있는 item을 리턴해야함을 주의

    def traverse(self):
        t_list = []
        
        curr_node = self.head  # 첫 노드에 접근
        
        while curr_node != None:  # tail이 되면 none이 됨
            t_list.append(curr_node.data)  # 노드가 아니라 item을 넣음
            curr_node = curr_node.next  # 다음 노드에 접근
            
        return t_list
```

### 4) 원소 삽입
0. prev(삽입 직전 노드) 찾기
1. newNode의 next link가 다음 위치(pos)를 가리키도록 지정
2. newNode를 pos-1의 link가 가리키도록 함
3. Nodecount를 1 증가시킴

![제목](캡처.jpg)

```python
# pos 위치에 삽입
# 이때 pos는 1<=pos<=nodeCount+1 범위를 가짐

def insertAt(self, pos, newNode):
    prev = self.getAt(pos-1)  # prev : 삽입 직전 노드
    newNode.next = prev.next
    prev.next = nextNode
    self.nodeCount += 1
```

```python
# newNode를 삽입하고 성공/실패에 따라 T/F 리턴하기!
def insertAt(self, pos, newNode):
    
    if pos<1 or pos>self.nodeCount:
        return False   # 함수 종료
    
    # 리스트 맨 앞 삽입(예외1)
    if pos==1: 
        newNode.next = self.head
        self.head = newNode
      
    else:
        prev = self.getAt(pos-1)  # prev : 삽입 직전 노드 찾기
        newNode.next = prev.next  # 기존 prev의 next를 가리키도록 함
        prev.next = nextNode  # 이제 기존 prev의 next는 nextNode로
        
        
    # 리스트 맨 뒤 삽입(예외2)
    if pos == self.nodeCount + 1:
        self.tail = newNode
        
    self.nodeCount += 1
    return True
```

### 6) 두 리스트의 연결

```python
# L : 이어붙일 리스트
def concat(self, L):
    self.tail.next = L.head  # 원래 리스트의 맨 끝이 이어붙일 리스트의 처음
    
    # L에 주어진 리스트가 빈 리스트일 때 예외처리
    if L.tail:
        self.tail = L.tail  # 내 원래 tail을 이어붙인 리스트의 tail로 가게함
        
    self.nodeCount += L.nodeCount # count는 두개를 합한 만큼
    
```
