# 8. 양방향 연결리스트

- 한쪽으로만 링크를 연결하지 말고, 양쪽으로 링크를 연결하자!
> 앞으로도(다음 node), 뒤로도 (이전 node) 진행 가능

- Node의 구조를 확장한다! (prev, next)    

- 양쪽 노드가 있어서 메모리 공간은 많이 차지하지만, 그만큼 코딩하기는 편해진다!


```python
class Node:
    def __init__(self, item):
        self.data = item
        self.prev = None
        self.next = None  # 양방향으로 연결
```

- 리스트의 처음과 끝에 dummy node를 두자!
> 데이터를 담고있는 node들은 **모두 같은 모양**



```python
class DoublyLinkedList:
    def __init__(self, item):
        self.nodeCount = 0
        self.head = Node(None)      # dummy node를 만든다!
        self.tail = Node(None)
        
        # 연결리스트 본격 구현(연결 노드가 두개인 리스트)
        self.head.prev = None
        self.head.next = self.tail
        self.tail.prev = self.head
        self.tail.next = None
```

## 연결리스트 연산

### 1. 리스트 순회

```python
# 리스트에 원소 담기
def traverse(self):
    result = []
    curr = self.head  ## 빈 리스트에 대해서도 dummy node때문에 실행됨
    
    while curr.next.next:
        curr = curr.next
        result.append(curr.data)
        
    return result
```

### 2. 리스트 역순회

```python
def reverse(self):
    result = []
    curr = self.tail
    
    while curr.prev.prev:
        curr = curr.prev
        result.append(curr.data)
    return result
```

### 3. 원소의 삽입

```python
def insertAfter(self, prev, newNode):
    next = prev.next   # next node 지정
    newNode.prev = prev
    newNode.next = next
    prev.next = newNode
    next.prev = newNode
    
    self.nodeCount += 1
    return True
```

```python
def insertAt(self, pos, newNode):
    if pos<1 or pos>self.nodeCount +1:
        return False
    
    prev = self.getAt(pos-1)
    return self.insertAfter(prev, newNode)

## 리스트 마지막에 원소를 삽입하면??? 
# 앞에서부터 찾아가서 삽입을 해야함! >> get at 매서드를 수정해보자!
```

```python
## 수정된 getAt 매서드
def getAt(self, pos):
    if pos<0 or pos>self.nodeCount:
        return None
    
    # pos가 nodeCount보다 뒤쪽이면 뒤에서 하나하나 세서 오도록
    if pos> self.nodeCount//2:
        i=0
        curr = self.tail
        while i<self.nodeCount - pos + 1:
            curr = curr.prev
            i+=1
            
    # 앞쪽이면 원래 코드 그대로
    else:
        i=0
        curr = self.head
        while i<pos:
            curr = curr.next
            i+=1
        
    return curr
```
